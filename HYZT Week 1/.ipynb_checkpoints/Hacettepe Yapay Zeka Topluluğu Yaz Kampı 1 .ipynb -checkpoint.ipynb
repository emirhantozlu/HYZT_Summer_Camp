```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import numpy as np
```

```python
# Verileri yükleme
train_df = pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\Hacettepe Yapay Zeka Yaz Kampı\veri setleri\train_data.csv")
predict_df = pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\Hacettepe Yapay Zeka Yaz Kampı\veri setleri\predict_data.csv")
```

```python
# Eksik değerleri doldurma fonksiyonu
def fill_missing_values(df):
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column] = df[column].fillna(df[column].mode()[0])  # Kategorik sütunlar için mod ile doldurma
        else:
            df[column] = df[column].fillna(df[column].mean())  # Sayısal sütunlar için ortalama ile doldurma
    return df
```

```python
# Eksik değerleri doldurma
train_df = fill_missing_values(train_df)
predict_df = fill_missing_values(predict_df)
```

```python
# Kategorik değişkenleri one-hot encoding ile dönüştürme
train_df = pd.get_dummies(train_df)
predict_df = pd.get_dummies(predict_df)
```

```python
# Train ve predict dataframes aynı sütunlara sahip olmalı, bu yüzden predict_df'e eksik sütunları ekleyelim
missing_cols = set(train_df.columns) - set(predict_df.columns)
for col in missing_cols:
    predict_df[col] = 0
```

```python
# Sütunları aynı sıraya getirelim
predict_df = predict_df[train_df.columns.drop('price')]
```

```python
# Veriyi ölçeklendirme
scaler = StandardScaler()

X_train = train_df.drop('price', axis=1)
y_train = train_df['price']
X_train_scaled = scaler.fit_transform(X_train)
X_predict_scaled = scaler.transform(predict_df)
```

```python

# Veriyi eğitim ve test setlerine bölmek
X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X_train_scaled, y_train, test_size=0.2, random_state=42)
```

```python

# RandomForestRegressor modeli
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train_split, y_train_split)
```

```python

# GradientBoostingRegressor modeli
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train_split, y_train_split)
```

```python
# RandomForestRegressor modeli ile test seti üzerinde tahmin oluşturma
y_pred_rf = rf_model.predict(X_test_split)
rmse_rf = np.sqrt(mean_squared_error(y_test_split, y_pred_rf))
print("Random Forest Root Mean Squared Error:", rmse_rf)
```

```python
# GradientBoostingRegressor modeli ile test seti üzerinde tahmin oluşturma
y_pred_gb = gb_model.predict(X_test_split)
rmse_gb = np.sqrt(mean_squared_error(y_test_split, y_pred_gb))
print("Gradient Boosting Root Mean Squared Error:", rmse_gb)
```

```python
# Ensemble model oluşturma (örneğin, basit bir ortalama alma)
y_pred_ensemble = (y_pred_rf + y_pred_gb) / 2
rmse_ensemble = np.sqrt(mean_squared_error(y_test_split, y_pred_ensemble))
print("Ensemble Root Mean Squared Error:", rmse_ensemble)
```

```python
# Predict veri seti üzerinde tahminlerde bulunma
predictions_rf = rf_model.predict(X_predict_scaled)
predictions_gb = gb_model.predict(X_predict_scaled)
```

```python
# Ensemble tahminlerini oluşturma (örneğin, basit bir ortalama alma)
predictions_ensemble = (predictions_rf + predictions_gb) / 2
```

```python
# Örnek gönderim dosyasını oluşturma
submission_ensemble = pd.DataFrame({
    "ID": predict_df.index,
    "Price": predictions_ensemble
})

submission_ensemble.to_csv('submission.csv', index=False)
```
