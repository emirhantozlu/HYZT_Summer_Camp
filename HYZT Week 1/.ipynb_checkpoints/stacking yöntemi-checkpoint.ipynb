```python
# Gerekli kütüphaneleri import etme
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from xgboost import XGBRegressor
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Verileri yükleme
train_df = pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\Hacettepe Yapay Zeka Yaz Kampı\Hacettepe Yapay Zeka Yaz Kampı Hafta 1\veri setleri\train_data.csv")
predict_df = pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\Hacettepe Yapay Zeka Yaz Kampı\Hacettepe Yapay Zeka Yaz Kampı Hafta 1\veri setleri\predict_data.csv")
```

```python
# Eksik değerleri doldurma
def fill_missing_values(df):
    for column in df.columns:
        if df[column].isnull().sum() > 0:
            df[column].fillna(df[column].mean(), inplace=True)
    return df

train_df = fill_missing_values(train_df)
predict_df = fill_missing_values(predict_df)
```

```python
# Aykırı değerleri kontrol etme ve kaldırma
plt.figure(figsize=(10, 6))
sns.boxplot(x=train_df['price'])
plt.show()

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.01)
    Q3 = df[column].quantile(0.99)
    IQR = Q3 - Q1
    df = df[(df[column] >= Q1) & (df[column] <= Q3)]
    return df

train_df = remove_outliers(train_df, 'price')
```

```python
# Yeni değişkenler oluşturma
train_df['age'] = 2024 - train_df['yr_built']
predict_df['age'] = 2024 - predict_df['yr_built']
```

```python
# Kategorik değişkenleri one-hot encoding ile dönüştürme
train_df = pd.get_dummies(train_df)
predict_df = pd.get_dummies(predict_df)
```

```python
# Train ve predict dataframes aynı sütunlara sahip olmalı, bu yüzden predict_df'e eksik sütunları ekleyelim
missing_cols = set(train_df.columns) - set(predict_df.columns)
for col in missing_cols:
    predict_df[col] = 0
```

```python
# Sütunları aynı sıraya getirelim
predict_df = predict_df[train_df.columns.drop('price')]
```

```python
# Veriyi ölçeklendirme
scaler = StandardScaler()
X_train = train_df.drop('price', axis=1)
y_train = train_df['price']
X_train_scaled = scaler.fit_transform(X_train)
X_predict_scaled = scaler.transform(predict_df)
```

```python
# Veriyi eğitim ve test setlerine bölmek
X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X_train_scaled, y_train, test_size=0.2, random_state=42)
```

```python
# Random Forest ve Gradient Boosting için hiperparametre arama
```

```python
rf_model = RandomForestRegressor(random_state=42)
rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
rf_random_search = RandomizedSearchCV(rf_model, param_distributions=rf_params, n_iter=50, cv=3, verbose=2, random_state=42, n_jobs=-1)
rf_random_search.fit(X_train_split, y_train_split)
print("Random Forest Best Params:", rf_random_search.best_params_)
```

```python
gb_model = GradientBoostingRegressor(random_state=42)
gb_params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.05],
    'max_depth': [3, 4, 5, 6],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 1.0]
}
gb_random_search = RandomizedSearchCV(gb_model, param_distributions=gb_params, n_iter=50, cv=3, verbose=2, random_state=42, n_jobs=-1)
gb_random_search.fit(X_train_split, y_train_split)
print("Gradient Boosting Best Params:", gb_random_search.best_params_)
```

```python
# Stacking Regressor modeli
xgb_model = XGBRegressor(random_state=42)
estimators = [
    ('rf', rf_random_search.best_estimator_),
    ('gb', gb_random_search.best_estimator_),
    ('xgb', xgb_model)
]
stacking_model = StackingRegressor(estimators=estimators, final_estimator=RandomForestRegressor(n_estimators=50, random_state=42))
stacking_model.fit(X_train_split, y_train_split)
```

```python
# Modellerin performans değerlendirmesi
y_pred_rf = rf_random_search.best_estimator_.predict(X_test_split)
rmse_rf = np.sqrt(mean_squared_error(y_test_split, y_pred_rf))
print("Random Forest Root Mean Squared Error:", rmse_rf)

y_pred_gb = gb_random_search.best_estimator_.predict(X_test_split)
rmse_gb = np.sqrt(mean_squared_error(y_test_split, y_pred_gb))
print("Gradient Boosting Root Mean Squared Error:", rmse_gb)

y_pred_stacking = stacking_model.predict(X_test_split)
rmse_stacking = np.sqrt(mean_squared_error(y_test_split, y_pred_stacking))
print("Stacking Model Root Mean Squared Error:", rmse_stacking)
```

```python
# Predict veri seti üzerinde tahminlerde bulunma
predictions_stacking = stacking_model.predict(X_predict_scaled)
```

```python
# Örnek gönderim dosyasını oluşturma
submission_stacking = pd.DataFrame({
    "ID": predict_df.index,
    "Price": predictions_stacking
})

submission_stacking.to_csv('submission_stacking.csv', index=False)
```
